# 1.0 Define functions

Firing rate equation for a single node

$$\tau_1 \frac{dx_1}{dt} = -x_1 + f\Big( w_{11} x_1 + b_1 + s_1 \Big) \qquad \qquad (1)$$


Where $f(x)$ refers to the nonlinear transfer function, $\tau_i$ refers to the time constant of neuron $i$, and $s_i$ is the gain on neuron $i$.


### 2.1 - Phase Portraits and Nullclines for $x_1$ 

Mathematical formula for $x_1$ nullcline:
$$ 0 = w_{11} x_1 + b_1 + s_1 - f^{-1}(x_1) $$



```python
def transfer(x,function='sigmoid'):
    if function=='sigmoid':
#         return 1/(1+np.exp(-.5*x))
        return 1/(1+np.exp(-x))

def dxdtSigmoid(x):
#     return np.exp(x)/((np.exp(x)+1)**2)
    return np.exp(x)/((np.exp(x)+1))

def invsigmoid(x):
#     return 2*np.log((x)/(1-(x)))
    return np.log((x)/(1-(x)))


def runSim1D(w11,d,b,s,t,xinit1,tau,dt=.1,function='sigmoid'):
    I1 = np.zeros((t,))
    # Set initial conditions
    I1[0] = xinit1
    
    for i in range(1,t):
        i1k1 = -I1[i-1] + transfer(w11*I1[i-1]+b+d[i-1]+s[i-1],function=function) 
        i1k1 = i1k1/tau
        
        i1a = I1[i-1] + i1k1*dt
        
        i1k2 = -i1a + transfer(w11*i1a+b+d[i]+s[i],function=function)
        i1k2 = i1k2/tau
        
        I1[i] = I1[i-1] + dt*((i1k1+i1k2)/2)
        
    return I1[25:]

def findFP(w11,b,s,t,xinit1,tau,dt=.1):
    """
    Finds the fixed point of the model for a particular parameter regime
    """
    I1 = np.zeros((t,))
    # Set initial conditions
    I1[0] = xinit1
    
    for i in range(1,t):
        i1k1 = -I1[i-1] + transfer(w11*I1[i-1]+b+s) 
        i1k1 = i1k1/tau
        
        i1a = I1[i-1] + i1k1*dt
        
        i1k2 = -i1a + transfer(w11*i1a+b+s)
        i1k2 = i1k2/tau
        
        I1[i] = I1[i-1] + dt*((i1k1+i1k2)/2)
        
        if I1[i] == I1[i-1] == I1[i-2]: 
            I1fp = I1[i] 
            break        
        
    return I1fp


def nullclineI1(w11,d1,I1,b1,s1,function='sigmoid'):
    if function=='sigmoid':
        pos_ind = np.where(I1>0)[0]
        invx1 = invsigmoid(I1[pos_ind])
        nI1 = (w11*I1[pos_ind] + b + s - invsigmoid(I1[pos_ind]))
        I1 = I1[pos_ind]
        return nI1, I1
    
```

```python
b = 0 # For fixed point = 0.5
x = np.arange(-5,5,.1)
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
plt.figure(figsize=(7,5))
plt.plot(x,transfer(x+b,function='sigmoid'),'k',linewidth=5)
plt.title('Sigmoid transfer function',fontsize=24)
plt.ylabel('Output',fontsize=20)
plt.xlabel('Input (a.u.)',fontsize=20)
plt.ylim([-.15,1.1])
plt.tight_layout()
# plt.savefig('Analysis0d_SigmoidFunction.pdf')

def normal(x,mean=0,sd=1):
    a = (1/(np.sqrt(2*np.pi*sd**2)))*np.exp(-(x-mean)**2/(2*sd**2))
    return a

x = np.arange(-5,5,.01)
y = normal(x)
plt.figure()
plt.title('Gaussian distribution (input)',fontsize=24)
plt.plot(x,y,'k')
# plt.savefig('Analysis0d_NormalDistribution_Input.pdf')

def SSN(x,bias):
    x = x + bias
    thresh = x > 0
    k = 0.04
    n = 2.0
#     y = np.multiply(k*(x)**n,thresh)
    y = np.log(1+np.exp(x))
#     if not excitatory:
#         y = -1.0*y
    return y

    


x = np.arange(-5,5,.01)
plt.figure(figsize=(7,5))
y = SSN(x,bias=3)
y = y/np.max(y)
plt.title('Sigmoid activation function emerges\nfrom balanced E-I populations', fontsize=24)
plt.plot(x,y,'#ff6666',label='$E(x)=log(1 + e^{x+b})$',linewidth=4)
y = -SSN(x,bias=-3)/np.min(y)
y = y/np.min(y)
plt.plot(x,y,'#8c8cd9',label='$I(x)=log(1 + e^{x-b})$',linewidth=4)
y = (SSN(x,bias=+3) - SSN(x,bias=-3))/np.max(y)
y = y/np.max(y)
plt.plot(x,y,'k--',label='$F(x) = E(x) - I(x)$',linewidth=8)
plt.ylabel('Normalized output',fontsize=20)
plt.xlabel('Input',fontsize=20)
plt.legend(prop={'size': 18})

plt.tight_layout()
# plt.savefig('Analysis0e_sigmoidValidation.pdf')



```

```python
##### Set connectivity weights
w11 = 1

# Set time constant
tau = .1

# Set n timepoints
t = 2025

d = np.random.normal(0,0.25,(t,))
stimrange = [-3,0,3]
I1fp = []
I1 = []
nI1 = []
sd_vals = []
for s in stimrange:
    s1vec = np.ones((t,))*s

    # Calculate fixed points
    xinit1 = 0
    I1fp.append(findFP(w11,b,s,10000,xinit1,tau))

    # Set initial conditions
    xinit1 = I1fp[-1]

    #####
    ## Run simulations
    I1.append(runSim1D(w11,d,b,s1vec,t,xinit1,tau))
    # Calculate variance
    sd_vals.append(np.var(I1[-1]))

    #####
    ## Calculate nullclines
    I1_sort = np.arange(0,1,.001)

    ntmp,xI1 = nullclineI1(w11,0,I1_sort,b,s)
    nI1.append(ntmp)

#####
# Generate plots
plt.figure(figsize=(10,7))
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
# Plot
plt.plot(range(len(I1[0])),I1[0],label='Evoked de-activation | Var = ' + str(round(sd_vals[0],7)),color='#333399',linewidth=4)
plt.plot(range(len(I1[1])),I1[1],label='Baseline activity | Var = ' + str(round(sd_vals[1],7)),color='k',linewidth=4)
plt.plot(range(len(I1[2])),I1[2],label='Evoked activation | Var = ' + str(round(sd_vals[2],7)),color='#800000',linewidth=4)
plt.xlabel('Time (s)',fontsize=30, fontname='FreeSans')
plt.ylabel('Activity',fontsize=30, fontname='FreeSans')
plt.title('Simulated neural activity during\ndifferent states', fontname='FreeSans', fontsize=35,y=1.02)
plt.xticks(np.arange(0,len(I1[0])+250,250),np.arange(0,len(I1[0])+250,250)*.01)
plt.ylim([0,1])
plt.legend(loc=1,bbox_to_anchor=(1.0,0.97), prop={'size': 24})
plt.tight_layout()
# plt.savefig('Analysis0e_simulatedTimeseries.pdf')

# Generate BOLD plots
BOLD = []
z = np.vstack((I1[0],I1[1],I1[2]))
BOLD = bw.balloonWindkessel(z,sampling_rate=0.01)[0]
plt.figure(figsize=(10,7))
plt.rc('xtick', labelsize=18)
plt.rc('ytick', labelsize=18)
# Plot
plt.plot(range(len(BOLD[0,:])),BOLD[0,:]-np.mean(BOLD[0,:]),
         label='Evoked de-activation',color='#333399',linewidth=4)
plt.plot(range(len(BOLD[1,:])),BOLD[1,:]-np.mean(BOLD[1,:]),
         label='Baseline activity',color='k',linewidth=4)
plt.plot(range(len(BOLD[2,:])),BOLD[2,:]-np.mean(BOLD[2,:]),
         label='Evoked activation',color='#800000',linewidth=4)
plt.xlabel('Time (s)',fontsize=30, fontname='FreeSans')
plt.ylabel('BOLD activity',fontsize=30, fontname='FreeSans')
plt.title('Simulated BOLD activity during\ndifferent states', fontsize=35,y=1.02)
plt.xticks(np.arange(0,len(I1[0])+250,250),np.arange(0,len(I1[0])+250,250)*.01)
# plt.legend(loc=1,bbox_to_anchor=(1.1,0.97), prop={'size': 24})
plt.tight_layout()
# plt.savefig('Analysis0e_BOLDTimeseries.pdf')


fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
plt.title('Phase space across\ndifferent states',fontsize=35, y=1.02,)
ax.plot(I1_sort[1:],nI1[0],alpha=0.9,label='$\dot{X_1}$ for evoked de-activation | ' + 'Fixed point = ' + str(round(I1fp[0],3)),color='#333399',linewidth=4)
ax.plot(I1fp[0],0,'*',color='#333399',markersize=15)

ax.plot(I1_sort[1:],nI1[1],alpha=0.9,label='$\dot{X_1}$ for baseline activity | ' + 'Fixed point = ' + str(round(I1fp[1],3)),color='k',linewidth=4)
ax.plot(I1fp[1],0,'*k',markersize=15)

ax.plot(I1_sort[1:],nI1[2],alpha=0.9,label='$\dot{X_1}$ for evoked activation | ' + 'Fixed point = ' + str(round(I1fp[2],3)),color='#800000',linewidth=4)
ax.plot(I1fp[2],0,'*',color='#800000',markersize=15)

ax.plot(I1_sort[1:],np.repeat(0,len(I1_sort[1:])),alpha=.5,color='k')
# plt.plot(I1[0],np.repeat(0,len(I1[0])),'ob',alpha=0.1,label='De-activated state trajectory', linewidth=.5)
# plt.plot(I1[1],np.repeat(0,len(I1[1])),'og',alpha=0.1,label='Spontaneous state trajectory', linewidth=.5)
# plt.plot(I1[2],np.repeat(0,len(I1[2])),'or',alpha=0.1,label='Activated state trajectory', linewidth=.5)
    
plt.xlabel('$X_1$',fontsize=35)
plt.ylabel('$\dot{X_1}$',fontsize=35)
# lgd = ax.legend(bbox_to_anchor=(1.0, 1.0),prop={'size': 20})
lgd = ax.legend(loc=1,prop={'size': 17})
plt.tight_layout()


# plt.savefig('Analysis0e_phasePortraitsAcrossStates.pdf',bbox_extra_artists=(lgd,), bbox_inches='tight')
```

# 3.0 Run linear stability analysis on fixed points across a range of stimulations

```python
import pandas as pd
def linearStabilityAnalysis1D(xstar,w,b,s,tau):
    """
    Calculates the characteristic time scale given a point xstar for our 1 dimensional system
    """
    secondderiv = (-1 + dxdtSigmoid(w*xstar+b+s))/float(tau)
    characteristic_timescale = 1/np.abs(secondderiv)
    return characteristic_timescale

def dxdtSigmoid(x):
    return np.exp(-x)/((np.exp(-x)+1)**2)
```

```python
# Set connectivity weights
w11 = 1
# Set time constant
tau = .1
# Set n timepoints
t = 2000

stimAmps = np.arange(-5,5.01,.05)
sd_values = []
entropy_values = []
kurtosis_values = []
sd_bold = []
timescale_values = []
tmp3 = []
for s in stimAmps:
    d = np.random.normal(0,0.25,(t,))
    s1vec = np.ones((t,))*s
    ## Calculate fixed point
    xinit1 = 0
    I1fp = findFP(w11,b,s,10000,xinit1,tau)
    tmp3.append(I1fp)
    # Set initial conditions
    xinit = I1fp
    ## Run simulations
    I1 = runSim1D(w11,d,b,s1vec,t,xinit1,tau)
    # Calculate Var
    sd_values.append(np.var(I1))
    # Calculate entropy
    entropy_values.append(stats.entropy(np.histogram(I1,bins=10)[0]))
    # Calculate kurtosis
    kurtosis_values.append(stats.kurtosis(I1))
    # Transform to BOLD
    BOLD = bw.balloonWindkessel(I1,sampling_rate=0.01)[0]
    sd_bold.append(np.var(BOLD))
    

    # Calculate characteristic time point
    timescale_values.append(linearStabilityAnalysis1D(I1fp,w11,b,s,tau))
```
